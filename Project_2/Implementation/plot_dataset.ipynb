# Plot Dataset P2

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
df_9832 = pd.read_csv("../Dataset/Almond9832.csv", header=None)
df_1212 = pd.read_csv("../Dataset/pecan1212.csv", header=None)
df_2621 = pd.read_csv("../Dataset/Walnut2621.csv", header=None)
```

```python
# plot all datasets
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
datasets = [
    (df_9832, "9832 Almonds"),
    (df_1212, "1212 Pecan"),
    (df_2621, "2621 Walnut"),
]

for idx, (df, title) in enumerate(datasets):
    ax = axes[idx]
    ax.scatter(df.iloc[1:, 0], df.iloc[1:, 1], color='blue', s=1, label='Data Points')
    ax.set_title(title)
    ax.set_xlabel('X Coordinate')
    ax.set_ylabel('Y Coordinate')
    ax.axis('equal')
    ax.grid(True)
    
plt.tight_layout()
plt.show()
```
