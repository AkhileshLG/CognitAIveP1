# Plot Dataset P2

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
df_9832 = pd.read_csv("../Dataset_csv/Almond9832.csv", header=None)
df_1212 = pd.read_csv("../Dataset_csv/pecan1212.csv", header=None)
df_2621 = pd.read_csv("../Dataset_csv/Walnut2621.csv", header=None)
```

```python
# plot all datasets
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
datasets = [
    (df_9832, "9832 Almonds"),
    (df_1212, "1212 Pecan"),
    (df_2621, "2621 Walnut"),
]

for idx, (df, title) in enumerate(datasets):
    ax = axes[idx]
    # Plot all points except the first in blue
    ax.scatter(df[0][1:], df[1][1:], c='blue', marker='o')
    # Plot the first point in orange
    ax.scatter(df[0][0], df[1][0], c='orange', marker='o')
    ax.set_title(title)
    ax.set_xlabel('X Coordinate')
    ax.set_ylabel('Y Coordinate')
    ax.axis('equal')
    ax.grid(True)
    
plt.tight_layout()
plt.show()
```
