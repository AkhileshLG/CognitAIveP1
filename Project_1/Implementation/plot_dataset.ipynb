# Plot Dataset P1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
df_32 = pd.read_csv("../Dataset_csv/32Almonds.csv", header=None)
df_64 = pd.read_csv("../Dataset_csv/64Walnut.csv", header=None)
df_128 = pd.read_csv("../Dataset_csv/128Hazelnut.csv", header=None)
df_128_circle = pd.read_csv("../Dataset_csv/128Circle201.csv", header=None)
df_256 = pd.read_csv("../Dataset_csv/256Cashew.csv", header=None)
```

```python
# check if data are clean: should contain all numbers
def check_data(file):
    try:
        df = pd.read_csv(file, header=None)
        df = df.apply(pd.to_numeric)
        print(f"Data in {file} is clean.")
    except ValueError:
        print(f"Data in {file} contains non-numeric values.")
        
check_data("../Dataset_csv/32Almonds.csv")
check_data("../Dataset_csv/64Walnut.csv")
check_data("../Dataset_csv/128Hazelnut.csv")
check_data("../Dataset_csv/128Circle201.csv")
check_data("../Dataset_csv/256Cashew.csv")
```

```python
# plot all datasets
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
datasets = [
    (df_32, "32 Almonds"),
    (df_64, "64 Walnut"),
    (df_128, "128 Hazelnut"),
    (df_128_circle, "128 Circle201"),
    (df_256, "256 Cashew")
]

for idx, (df, title) in enumerate(datasets):
    ax = axes[idx // 3, idx % 3]
    # Plot all points except the first in blue
    ax.scatter(df[0][1:], df[1][1:], c='blue', marker='o')
    # Plot the first point in orange
    ax.scatter(df[0][0], df[1][0], c='orange', marker='o')
    ax.set_title(title)
    ax.set_xlabel('X Coordinate')
    ax.set_ylabel('Y Coordinate')
    ax.axis('equal')
    ax.grid(True)

# Hide empty subplot
axes[1, 2].axis('off')
    
plt.tight_layout()
plt.show()
```
