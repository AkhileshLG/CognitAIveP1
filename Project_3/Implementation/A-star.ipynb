```python
# Store mainfest: NAN as -1, 0 as empty, >0 as occupied, number are weight of the container
def create_grid(path):
    grid = pd.DataFrame(0, index=range(0, 9), columns=range(1, 13))
    df = pd.read_csv(path)
    
    # flip the row
    max_row = df["row"].max()

    for _, rec in df.iterrows():
        r = max_row - int(rec["row"]) + 1
        c = int(rec["column"])
        desc = str(rec["description"]).strip()

        if desc == "NAN":
            grid.loc[r, c] = -1
        elif desc == "UNUSED":
            grid.loc[r, c] = 0
        else:
            grid.loc[r, c] = int(rec["weight"])

    return grid

grid6 = create_grid("../Dataset/ShipCase6.csv")
grid6
```

2. Define goal state

```python
def left_weight(grid):
    return int(grid.iloc[1:9, 0:6].clip(lower=0).to_numpy().sum())

def right_weight(grid):
    return int(grid.iloc[1:9, 6:12].clip(lower=0).to_numpy().sum())

def goal_test(grid, tol=0.10):
    total = left_weight(grid) + right_weight(grid)
    return True if total == 0 else abs(left_weight(grid) - right_weight(grid)) <= tol * total

goal_test(grid6)
```
